```python
data = []
for code in range(174850,174950):
    res = requests.get(url.format(code))
    soup = BeautifulSoup(re.sub("&#(?![0-9])","",res.text),'html5lib')
    body = soup.find('body')
    story1 = body.find('div',{'class':'story_area'}) 
    ko1 = hangle(story1)
    info = body.select('dd > p > span >a')
    ko2 = hangle(info)
    title = body.find('h3',{'class':'h_movie'}).findAll('a')[0]
    ko0 = hangle(title)
    people = body.select('dl.info_spec > dd > p')[1:3]
    ko3 = hangle(people)
    data.append([code,ko0,ko1,ko2,ko3])
    df = pd.DataFrame(data, columns =['code','제목','줄거리','개요','출연'])
    
df
```

```python
def hangle(s):
    return re.sub('[^ ㄱ-ㅣ가-힣]+','', str(s)).strip()
             
```
